```python
import re

# Clase Lectura
class Lectura:

    grados_f = 0.0

    def _init_(self, grados_f):
      self.grados_f = grados_f

    @property
    def grados_c(self):
      return (self.grados_f-32)/1.8

    @property
    def estado(self):
      if self.grados_c < 35:
        return "Hipotermia"
      if self.grados_c < 37:
        return "Normal"
      if self.grados_c < 38:
        return "Febrícula "
      if self.grados_c < 41:
        return "Fiebre"
      if self.grados_c > 40:
        return "Hiperpirexia"

    def mostrar_lectura(self):
      print("-" * 50)
      print(f'Farenheit: {self.grados_f:.3f}')
      print(f'Centigrados: {self.grados_c:.3f}')
      print(f'Estado: {self.estado}')


# Función capturar_f
def capturar_f(prompt='Dame un número flotante:'):
      regex = r"([0-9]{1,3}\.[0-9]{1,3})$"

      while True:
        try:
          prompt = input("\nIngresa la temperatura del paciente\n")
          temp = float(prompt)
        except:
          print('Dato no numérico. Ingresa un número.')
          continue

        if not (temp >= 0.0 and temp <= 150.0):
          print('Rango no aceptado. Ingresa un valor entre 0.0 y 150.0.')
          continue

        if not (re.match(regex, prompt)):
          print('Patrón no cumplido. El dato debe apegarse al patrón de uno a tres dígitos, un punto, y de una a tres decimal.')
          continue

        break

      return temp


# Programa principal
lecturas = []

print('Sistema de conversión de C° a F° y recopilación de temperaturas')
print('*El dato debe tener de uno a tres dígitos, un punto, y de una a tres decimales')
print('*El valor convertido a float debe estar entre 0.0 y 150.0')
while True:
      lectura = Lectura(capturar_f())
      lecturas.append(lectura)
      op = input('¿Deseas ingresar otra temperatura? Si/No \n')
      if op.upper() == 'NO':
        break
      elif op.upper() == 'SI':
        continue


# Impresion de lecturas
print(f'\nTotal de lecturas registradas: {len(lecturas)}')

i = 1
for lectura in lecturas:
  lectura.mostrar_lectura()
  print(f'\nLectura {i}')

  i += 1
```
